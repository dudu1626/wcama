```python
# Carregar os dados
nome_arquivo = 'emissoes_treino_total_unidades.csv'
dados_emissoes = pd.read_csv(nome_arquivo, sep=';', encoding='utf-8')

# Colunas numéricas a serem convertidas
colunas_numericas = [
    'duration', 'emissions', 'emissions_rate', 'cpu_power', 'gpu_power',
    'cpu_energy', 'gpu_energy', 'ram_energy', 'energy_consumed', 'carbono_g',
    'energia_consumida_wh', 'energia_W',
]

# Converter as colunas numéricas (trocando ',' por '.')
for coluna in colunas_numericas:
    dados_emissoes[coluna] = dados_emissoes[coluna].str.replace(',', '.').astype(float)


resultados_testes = []

for arquitetura in dados_emissoes['Arquitetura'].unique():

    # Filtrar os dados para a arquitetura atual
    dados_filtrados = dados_emissoes[dados_emissoes['Arquitetura'] == arquitetura]

    # Separar os dados por precisão
    fp32 = dados_filtrados[dados_filtrados['Precisão'] == 'FP32']
    fp16 = dados_filtrados[dados_filtrados['Precisão'] == 'FP16']

    for coluna in colunas_numericas:
        # Teste de normalidade (Shapiro-Wilk)
        stat_fp32, p_fp32 = shapiro(fp32[coluna])
        stat_fp16, p_fp16 = shapiro(fp16[coluna])

        normal_fp32 = p_fp32 > 0.05
        normal_fp16 = p_fp16 > 0.05

        # Escolha do teste
        if normal_fp32 and normal_fp16:

            stat, p_valor = ttest_rel(fp32[coluna], fp16[coluna])
            teste = 't-test pareado'
        else:
            stat, p_valor = wilcoxon(fp32[coluna], fp16[coluna])
            teste = 'Wilcoxon pareado'


        resultados_testes.append({
            'Arquitetura': arquitetura,
            'Métrica': coluna,
            'Teste': teste,
            'p-valor': p_valor,
            'Normal FP32': normal_fp32,
            'Normal FP16': normal_fp16,
    })

# Criar DataFrame com os resultados
resultados_df = pd.DataFrame(resultados_testes)

resultados_df['Significativo 5%'] = resultados_df['p-valor'] < 0.05
resultados_df['Significativo 5%'] = resultados_df['Significativo 5%'].replace({True: 'Sim', False: 'Não'})
resultados_df['Significativo 6%'] = resultados_df['p-valor'] < 0.06
resultados_df['Significativo 6%'] = resultados_df['Significativo 6%'].replace({True: 'Sim', False: 'Não'})
resultados_df['Normal FP32'] = resultados_df['Normal FP32'].replace({True: 'Sim', False: 'Não'})
resultados_df['Normal FP16'] = resultados_df['Normal FP16'].replace({True: 'Sim', False: 'Não'})

# salvar os resultados em arquivo excel
resultados_df.to_excel('resultados_emissoes_hipoteses_TREINO.xlsx', index=False)
```
